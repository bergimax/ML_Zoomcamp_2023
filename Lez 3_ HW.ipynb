# QUESTION 3

```python
df['above_average'] = (df.price >= df.price.mean()).astype(int)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42) #split in train e test, imposto seed=1
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42) #split in train e test, imposto seed=1
```

```python
len(df_train), len(df_test), len(df_val)
```

```python
y_train = df_train.above_average.values
y_val   = df_val.above_average.values
y_test  = df_test.above_average.values
```

```python
del df_train['above_average']
del df_val['above_average']
del df_test['above_average']
```

```python
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual_info_score_avg(series):
    return mutual_info_score(series, y_train)
```

```python
categorical = ['model', 'year', 'engine_hp',
               'engine_cylinders', 'transmission_type', 
               'vehicle_size', 'highway_mpg', 'city_mpg'
]
```

```python
mi = df_train[categorical].apply(mutual_info_score_avg) #apply the function to each column
mi = mi.sort_values(ascending=False)
round(mi, 2)
```

# QUESTION 4

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False) # transform a dictionary into a vector
```

```python
dicts = df_train[categorical].to_dict(orient='records')
dicts[0]
```

```python
dv.fit(dicts)
dv.get_feature_names_out()
```

```python
dv.transform(dicts)
```

```python
X_train = dv.fit_transform(dicts)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_[0].round(3)
```

```python
model.intercept_
```

```python
model.predict(X_train) #val of churn
```

```python
model.predict_proba(X_train) 
```

```python
val_dicts = df_val.to_dict(orient='records')
```

```python
X_val = dv.transform(val_dicts)
```

```python
y_pred = model.predict_proba(X_val)[:,1]
y_pred >= 0.5 
```

```python
decision = (y_pred >= 0.5)
decision
```

```python
acc = (y_val == decision).mean() #shortcut of what is below
round(acc, 2)
```
